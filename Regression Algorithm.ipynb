# Regression


```python
import statsmodels.api as sm
import pandas as pd
```

```python
sales_table= pd.read_excel(r'C:\Users\fast.LAB6-PC4\Desktop\Adventure Works 2020 (1).xlsx', sheet_name= 'Sales')
```

```python
sales_table
```

```python
sales_table.columns
```

```python
### Dependent Variable---------> Sales Amount
### Independent Variable-------> TotalProductCost, OrderQuantity, TaxAmt, SalesTerritory, UnitPrice
```

```python
dependent_variable=sales_table['SalesAmount']
independent_variable=sales_table[['TotalProductCost', 'OrderQuantity', 'TaxAmt','SalesTerritoryKey','UnitPrice']]
```

```python
dependent_variable
```

```python
independent_variable
```

```python
# Multiple linear Regression
```

```python
regression_model=sm.OLS(dependent_variable,independent_variable)
results=regression_model.fit()
results.summary()
```

```python
# Interpretation 

#P value should be less than 0.05
#P value of Unit Price is less than 0.05 ------>

#- ve relations More tax amount less Sales Amount
#               Less Tax Amount More Sales Amount
```
